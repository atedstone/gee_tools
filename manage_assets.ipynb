```python
import ee
ee.Initialize()
```

```python
assets = ee.data.listAssets({'parent':'projects/firn-retention/assets'})
```

```python
asset_ids = [a['id'] for a in assets['assets']]
```

```python
import re
```

```python
# ref_IW_HV_99_2019_2022_61_106
#match_pattern = r'ref_IW_HV_[0-9]{1,3}_[0-9]{4}_[0-9]{4}_[0-9]{1,3}_[0-9]{2,3}_100m_lee5$'
#match_pattern = r'wetness_accumz([a-zA-Z0-9\_])\w+_gris$'
match_pattern = r'ref_IW_HV_[0-9]{1,3}_2017_2018_[0-9]{1,3}_[0-9]{2,3}_[0-9]{2,3}m$'
#match_pattern = r'ref_IW_HV_[0-9]{1,3}_([0-9]{4})_\1_[0-9]{1,3}_[0-9]{2,3}' #With backreference to capture single-year assets

queue = []
for asset_id in asset_ids:
    if re.match(match_pattern, asset_id.split('/')[-1]) is not None:
        queue.append(asset_id)
```

```python
queue
```

```python
for asset_id in queue:
    #ee.data.renameAsset(asset_id, asset_id + '_BADBORDERS')
    ee.data.deleteAsset(asset_id)
```
